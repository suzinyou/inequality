```python
def get_truncated_normal(mean=0, sd=1, low=0, upp=10):
    return scipy.stats.truncnorm(
        (low - mean) / sd, (upp - mean) / sd, loc=mean, scale=sd)
```

```python
def generate_yearly_income(row):
    yearly_income = []
    
    lifespan = np.random.randint(75, 100)
    count, bins = np.histogram(np.random.normal(size=1000), bins=lifespan)
    lifetime_income = count / count.max() * row.max_income
    
    for year in years:
        age = year - row.BYEAR
        if age <= 0 or age >= lifespan:
            yearly_income.append(None)
        else:
            yearly_income.append(lifetime_income[age])
    return yearly_income
```

```python
population_info['yearly_income'] = population_info.apply(generate_yearly_income, axis=1)
```

```python
N_workplaces = 6000
workplaces = np.random.randint(10000, 99999, size=N_workplaces)
num_workers = np.around(np.random.lognormal(1, 1.5, size=N_workplaces)).astype(int)
num_workers[num_workers < 1] = 1
workplace_code = np.random.choice(geocode.SIDO_SGG_EMD_CD, replace=False, size=N_workplaces)
```

가구별 주소 선택

```python
def select_area_from_sido(sidos):
    filter_mask = (geocode.SIDO_CD == sidos[0]) | (geocode.SIDO_CD == sidos[1]) | (geocode.SIDO_CD == sidos[2])
    return np.random.choice(geocode.SIDO_SGG_EMD_CD[filter_mask].tolist())
```

```python
dfs = dict()
        
for year in years:
    this_year_income = population_info.yearly_income.apply(lambda x: x[2017 - year])
    
    notnull_mask = this_year_income.notnull()
    df = population_info[notnull_mask].drop(['시도가능', '가구주가능', 'max_income', 'yearly_income'], axis=1)
    df['HHRR_HEAD_INDI_DSCM_NO_MY'] = population_info.loc[notnull_mask, '가구주가능'].apply(
        lambda x: np.random.choice(x, p=[0.6, 0.3, 0.1]))
    df['RVSN_ADDR_CD'] = population_info.loc[notnull_mask, '시도가능'].apply(select_area_from_sido)
    
    # INCOME
    df['INC_TOT'] = this_year_income[notnull_mask]
    split = np.random.random((df.shape[0], 7))
    weight = split / split.sum(axis=1)[:, np.newaxis]
    df['INC_WAGE'] = df.INC_TOT * weight[:, 0]
    df['INC_BUS'] = df.INC_TOT * weight[:, 1]
    df['INC_INT'] = df.INC_TOT * weight[:, 2]
    df['INC_DIVID'] = df.INC_TOT * weight[:, 3]
    df['INC_PNSN_NATL'] = df.INC_TOT * weight[:, 4]
    df['INC_PNSN_OCCUP'] = df.INC_TOT * weight[:, 5]
    df['INC_OTHR'] = df.INC_TOT * weight[:, 6]
    
    # PROPERTY TAXBASE
    split = np.random.random((df.shape[0], 5))
    weight = split / split.sum(axis=1)[:, np.newaxis]
    df['prop_txbs_tot'.upper()] = np.around(np.random.lognormal(16, 1, df.shape[0]))
    for i, txbs_col in enumerate(['prop_txbs_bldg', 'prop_txbs_lnd', 'prop_txbs_hs', 'prop_txbs_shop', ]):
        df[txbs_col.upper()] = df.PROP_TXBS_TOT * weight[:, i]
    
    # FIRM
    missing_mask = np.random.choice([True, False], size=df.shape[0], p=[0.2, 0.8])
    firm_indices = np.random.choice(np.arange(N_workplaces), p=num_workers / num_workers.sum(), size=df.shape[0])
    df['FIRM_CD'] = workplaces[firm_indices]
    # merge later: df['FIRM_SCL_ENTER_NOP_ID']
    df['RVSN_FIRM_ADDR_CD'] = workplace_code[firm_indices]
    df.loc[missing_mask, 'FIRM_CD'] = np.nan
    df.loc[missing_mask, 'RVSM_FIRM_ADDR_CD'] = np.nan
    firm_cnt = df['FIRM_CD'].value_counts()

    #직장가입자여부
    df['EMP_Y'] = 'Y'
    missing_mask_2 = np.random.choice([True, False], size=df.shape[0], p=[0.6, 0.4])
    df.loc[np.logical_or(missing_mask, missing_mask_2), 'EMP_Y'] = None
    
    df['STD_YYYY'] = year
    
    dfs[year] = df
```
