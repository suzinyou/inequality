```python
def get_truncated_normal(mean=0, sd=1, low=0, upp=10):
    return scipy.stats.truncnorm(
        (low - mean) / sd, (upp - mean) / sd, loc=mean, scale=sd)
```

```python
N = 10000
population_info = pd.DataFrame(
    {
        'INDI_DSCM_N': np.arange(N),
        'SEX_TYPE': np.random.randint(1, 3, size=N),
        'BYEAR': np.round(get_truncated_normal(mean=1969, sd=13, low=1900, upp=2003).rvs(N)).astype(int),
        'GAIBJA_TYPE': np.random.choice([1, 2, 5, 6, 7, 8], size=N),
        'FOREIGNER_Y': np.random.choice([' ', 'Y'], p=[0.99, 0.01], size=N),
        '시도가능': np.random.choice(np.arange(17), p=np.arange(17, 0, -1)/ np.sum(np.arange(1, 18)), size=(N,3)).tolist(),
        'max_income': np.random.lognormal(mean=15, sigma=0.75, size=N)
    }
)
population_info['가구주가능'] = population_info.INDI_DSCM_N.apply(lambda x: np.r_[x, np.random.randint(N, size=2)])
```

```python
N_per_year = 100
years = np.arange(2003, 2018)
```

```python
count, bins = np.histogram(np.random.normal(size=1000), bins=80)
```

```python
def generate_yearly_income(row):
    yearly_income = []
    
    lifespan = np.random.randint(75, 100)
    count, bins = np.histogram(np.random.normal(size=1000), bins=lifespan)
    lifetime_income = count / count.max() * row.max_income
    
    for year in years:
        age = year - row.BYEAR
        if age <= 0 or age >= lifespan:
            yearly_income.append(None)
        else:
            yearly_income.append(lifetime_income[age])
    return yearly_income
```

```python
population_info['yearly_income'] = population_info.apply(generate_yearly_income, axis=1)
```

```python
this_year_income = population_info.yearly_income.apply(lambda x: x[2017 - years[0]])
print("Not null: ", this_year_income.notnull().sum())
notnull_mask = this_year_income.notnull()
df = population_info[notnull_mask].drop(['가구주가능', 'max_income', 'yearly_income'], axis=1)
```

```python
df['HHRR_HEAD_INDI_DSCM_NO_MY'] = population_info['가구주가능'].apply(lambda x: np.random.choice(x, p=[0.6, 0.3, 0.1]))
df['INC_TOT'] = this_year_income[notnull_mask]
df.head()
```

```python
dfs = dict()
        
for year in years:
    this_year_income = population_info.yearly_income.apply(lambda x: x[2017 - year])
    
    notnull_mask = this_year_income.notnull()
    df = population_info[notnull_mask].drop(['시도가능', '가구주가능', 'max_income', 'yearly_income'], axis=1)
    df['HHRR_HEAD_INDI_DSCM_NO_MY'] = population_info.loc[notnull_mask, '가구주가능'].apply(
        lambda x: np.random.choice(x, p=[0.6, 0.3, 0.1]))
    df['RVSN_ADDR_CD'] = population_info.loc[notnull_mask, '시도가능'].apply(
        lambda x: np.random.choice(x, p=[0.8, 0.15, 0.05]))
    df['INC_TOT'] = this_year_income[notnull_mask]
    split = np.random.random((df.shape[0], 7))
    weight = split / split.sum(axis=1)[:, np.newaxis]
    df['INC_WAGE'] = df.INC_TOT * weight[:, 0]
    df['INC_BUS'] = df.INC_TOT * weight[:, 1]
    df['INC_INT'] = df.INC_TOT * weight[:, 2]
    df['INC_DIVID'] = df.INC_TOT * weight[:, 3]
    df['INC_PNSN_NATL'] = df.INC_TOT * weight[:, 4]
    df['INC_PNSN_OCCUP'] = df.INC_TOT * weight[:, 5]
    df['INC_OTHR'] = df.INC_TOT * weight[:, 6]
    df['STD_YYYY'] = year
    
    dfs[year] = df
```

```python
df.merge(df.groupby('HHRR_HEAD_INDI_DSCM_NO_MY').size().rename('CNT_ID_HHRR_FD'), on='HHRR_HEAD_INDI_DSCM_NO_MY')
```
